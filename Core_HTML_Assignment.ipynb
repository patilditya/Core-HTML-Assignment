```python
# Question 5:  Create a description list of full stack web development tech stack, using the <dl> tag. Each term should bea tech stack name and each description should be a brief explanation of what the tech stack is used for

-->
    <h1>Full Stack Web Development Tech Stack</h1>

    <dl>
        <dt>HTML (Hypertext Markup Language)</dt>
        <dd>HTML is the standard markup language for creating web pages. It structures content like headings, paragraphs, links, and images on the webpage.</dd>

        <dt>CSS (Cascading Style Sheets)</dt>
        <dd>CSS is used to style and layout web pages. It controls the appearance of HTML elements, such as colors, fonts, and layout.</dd>

        <dt>JavaScript</dt>
        <dd>JavaScript is a programming language that adds interactivity to web pages. It is used to create dynamic content like form validation, animations, and real-time updates.</dd>

        <dt>React</dt>
        <dd>React is a JavaScript library used for building user interfaces, particularly single-page applications. It allows developers to create reusable UI components.</dd>

        <dt>Node.js</dt>
        <dd>Node.js is a JavaScript runtime environment that enables JavaScript to be run on the server side. It is commonly used for building scalable backend applications.</dd>

        <dt>Express.js</dt>
        <dd>Express.js is a web application framework for Node.js. It simplifies the creation of server-side applications and APIs, providing a robust set of features for building web apps.</dd>

        <dt>MongoDB</dt>
        <dd>MongoDB is a NoSQL database that stores data in a flexible, JSON-like format. It is commonly used for handling large-scale, unstructured data.</dd>

        <dt>SQL (Structured Query Language)</dt>
        <dd>SQL is used to interact with relational databases. It allows developers to query, insert, update, and delete data stored in tables.</dd>

        <dt>Git</dt>
        <dd>Git is a version control system that tracks changes in code and allows developers to collaborate efficiently. It is commonly used with GitHub or GitLab for project management and code sharing.</dd>

        <dt>Docker</dt>
        <dd>Docker is a platform that enables developers to create, deploy, and run applications in containers. It ensures that software runs consistently across different environments.</dd>

        <dt>AWS (Amazon Web Services)</dt>
        <dd>AWS provides a suite of cloud services like computing power, storage, and databases. It is widely used for hosting web applications and scaling infrastructure.</dd>

        <dt>Vue.js</dt>
        <dd>Vue.js is a JavaScript framework used for building user interfaces. It is known for its simplicity and flexibility, allowing developers to create interactive front-end applications.</dd>
    </dl>
```
